```julia
canon = open("C://Users//Vladimir Alencar//Desktop//cano.txt")
```

```julia
wordlist = split(lowercase(readall(canon)), r"\W", false)
```

```julia
wordcounts = Dict{ASCIIString,Int64}()
```

```julia
for word in wordlist
    wordcounts[word]=get(wordcounts, word, 0) + 1
end
```

```julia
wordcounts
```

```julia
wordcounts["watson"], wordcounts["holmes"], wordcounts["sherlock"],wordcounts["lestrade"]
```

```julia
get(wordcounts, "crime", 0)
```

```julia
get(wordcounts, "mistery", 0)
```

```julia
for i in sort(collect(keys(wordcounts)))
    println("$i, $(wordcounts[i])") end
```

```julia
sort(collect(wordcounts), by = tuple -> last(tuple), rev=true)
```

```julia
sort(collect(wordcounts), by = tuple -> last(tuple), rev=true)[1:100]
```

```julia
filter(tuple -> startswith(first(tuple), "k") && last(tuple) < 4, collect(wordcounts))
```
